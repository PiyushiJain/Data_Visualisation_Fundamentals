```python
x=np.arange(7,47)
y=x**2
y1=x**3
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('MY GRAPH')
plt.grid()
plt.legend(['yellow','pink'])
plt.show()
```

```python
plt.figure(figsize=(4,8))
plt.subplot(2,1,1)
plt.plot(x,y,ls='-.',color='red',lw=7,marker='o',markersize=7,markerfacecolor='blue',markeredgecolor='black')
plt.subplot(2,1,2)
plt.plot(x,y1,ls='--')
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
#categorical plots using seaborn
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.factorplot(x='day',y='total_bill',kind='box',data=df)
```
